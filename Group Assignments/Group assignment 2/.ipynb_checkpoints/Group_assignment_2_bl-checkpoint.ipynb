Now, I'd like to make some maps to see what the spatial distribution of population density looks like. Because my current dataset does not have lat/long or a geometry column, I'll need to join it with a geojson file. I'll first download the library 'geopandas' so that I can make maps, and then I'll import a new [geojson file](https://opendata.dc.gov/datasets/DCGIS::census-tracts-in-2020/explore) that covers the 2020 census tracts for Washington D.C.

```python
import geopandas as gpd
```

```python
#read the geojson file with D.C. geometry census tract data
tracts=gpd.read_file('data/Census_Tracts_in_2020.geojson')
tracts.head()
```

```python
#plot the data
tracts.plot(figsize=(12,10))
```

```python
#find out more about the data
tracts.info(verbose=True, show_counts=True)
```

I'm going to try subsetting the data so that it only includes TRACT and geometry, and then use it to join with the previous population density census data.

```python
tracts = tracts[['TRACT','geometry']]
tracts.head()
```

I want to make a FIPS column to match my other dataset. So given the [FIPS ID information for Washington DC](https://dc.postcodebase.com/county/11001), I'll run the code below.

```python
#make a FIPS column
tracts['FIPS']='11'+'001'+ tracts['TRACT']
#check the data
tracts.head()
```

```python
#check data
tracts.info(verbose=True,show_counts=True)
```

I'll now try merging the data between the census population density file and the census tract geojson file.

```python
tracts_pop_den=tracts.merge(pop_den2,on="FIPS")
```

```python
#check data
tracts_pop_den.head()
```

It looks like I was able to join the data using the FIPS column successfully. I want to make some choropleth maps now. I will use different interval types to see what the maps look like.

```python
#plot and map population density, using natural breaks
tracts_pop_den.plot(figsize=(12,10),
                    column='Population Density',
                    legend= True,
                    scheme='NaturalBreaks')
```

```python
#plot and map population density using equal intervals
tracts_pop_den.plot(figsize=(12,10),
                    column='Population Density',
                    legend=True,
                    scheme='equal_interval')
```

```python
#plot and map population density using quatiles
tracts_pop_den.plot(figsize=(12,10),
                    column='Population Density',
                    legend=True,
                    scheme='quantiles')
```

It looks like along all three maps, the greatest population density is concentrated towards the center of Washington D.C. For future analysis, I'm looking forward to putting different food data markers on top of these maps. I'd like to see whether the areas of higher population density have food access proportionate to the larger number of residents there.
