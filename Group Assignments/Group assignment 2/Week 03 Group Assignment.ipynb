## Graphs and Maps
### Imported geopandas to start visualizing the income data in graphs and maps

```python


import geopandas as gpd
```

```python
df2.plot(figsize=(12,10))
```

### Downloaded geojson file with DC census tracts from the DC open data portal to begin mapping the data

```python

tracts=gpd.read_file('Census_Tracts_in_2020.geojson')
tracts.head()
```

```python
tracts.plot(figsize=(12,10))
```

```python
tracts.info(verbose=True, show_counts=True)
```

### Subsetting the data to only inlcude geometry and FIPS in order to merge and join the data

```python

tracts = tracts[['TRACT','geometry']]
tracts.head()
```

```python
tracts['FIPS'] ='11' + '001' + tracts['TRACT']
```

```python
tracts.head()
```

```python
tracts_income=tracts.merge(df2,on="FIPS")
```

```python
tracts_income.head()
```

### Created choropleth maps in geopandas to visualize income data
### Map of areas with incomes less than $10,000 per year in 2020

```python

tracts_income.plot(figsize=(12,10),
                 column='Less than $10,000',
                 legend=True, 
                 scheme='NaturalBreaks')
```

### Map of areas with incomes greater than $200,000 per year in 2020

```python

tracts_income.plot(figsize=(12,10),
                 column='$200,000 or More',
                 legend=True, 
                 scheme='NaturalBreaks')
```

```python
tracts_income['PCT_Under $10,000'] = tracts_income['Less than $10,000']/tracts_income['Number of Households']*100
tracts_income['PCT_Over $200,000'] = tracts_income['$200,000 or More']/tracts_income['Number of Households']*100
```

### Map of areas with incomes less than $10,000 per year in 2020, shown as percentages across census tracts


```python
tracts_income.plot(figsize=(12,10),
                 column='PCT_Under $10,000',
                 legend=True, 
                 scheme='equal_interval')
```

### Map of areas with incomes greater than $200,000 per year in 2020, shown as percentages across census tracts

```python
tracts_income.plot(figsize=(12,10),
                 column='PCT_Over $200,000',
                 legend=True, 
                 scheme='equal_interval')
```
